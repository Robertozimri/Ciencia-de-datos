```python
#d.copy se crea una copia del diccionario 
datos = {'Crossfox': 72832.16, 'DS5': 124549.07, 'Jetta Variant': 88078.64}
datos_copia = datos.copy()
```

```python
datos_copia
```

```python
#d.pop(key [, default]). Si se encuentra el ítem es eliminado y su valor es retornado y sino se encuentra el valor especificado como defaul
datos_copia.pop('DS5')
```

```python
datos_copia
```

```python
#Este no lo encuetra por eso mismo marca error

datos_copia.pop('Passat')
```

```python
datos_copia.pop('Passat', 'Clave no encintrada.')
```

```python
#d.get(key[, default]) retorna el valor correspodiente a la clave key 
datos.get('DS5')
```

```python
#En caso de que no se encuentra el metodo nos retornara None o sea que este metodo nunca nos retorna un error 
datos.get('DS5x')
```

```python
#3.8 Para saber más deepcopy es lo mismo cuando utilizamos copy
dataset = {
'Dodge': {
'motor': 'Motor 6.7',
'año': 2017,
'km': 45757,
'accesorios': ['LLantas de aleación', '4 X 4', 'Central multimedia'],
'valor': 92612.1
},
'Pajero': {
'motor': 'Motor 2.4 Turbo',
'Cero_km': True,
'km': 80000,
'accesorios': ['Control de tracción', 'Bancos de cuero', 'Control de estabilidad'],
'valor': 51606.59
}
}
```

```python
dataset_copia = dataset.copy()
dataset_copia
```

```python
dataset_copia['Dodge']['accesorios'].append('Airbang')
dataset_copia
```

```python
#El atributo copy no copia bojetos sino crea referencias  
#dataset
```

```python
import copy
```

```python
#Para una copia profunda se utiliza deepcopy
dataset_copia = copy.deepcopy(dataset)
```

```python
#Ya se pueden hacer alteraciones en el original 
dataset_copia['Dodge']['accesorios'].append('Bancos de cuero')
dataset_copia
```
