```python
import pandas as pd
import numpy as np
```

```python
proyecto=pd.read_csv('https://raw.githubusercontent.com/Robertozimri/Robertozimri/main/db_proyecto.csv', sep=';')
proyecto.head()
```

```python
proyecto['ValorTotal'] = proyecto['Valor'] + proyecto['ValorMantenimiento']
proyecto.head()
```

```python
proyecto.groupby(by=['Tipo']) [['ValorTotal','Area']].mean().round(2)
```

```python
valor_medio_barrio=proyecto.groupby(by=['Barrio']) [['ValorTotal','Area']].mean().round(2)
valor_medio_barrio
```

```python
barrios=['Carola', 'Del Valle', 'Condesa', 'Moctezuma', 'Santa Fe Cuajimalpa']
valor_medio_barrio.loc[barrios]
```
