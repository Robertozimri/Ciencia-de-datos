```python
#Aqui nos aparece una tabla que se llama km_ media con las mediadas de todo 
for index, row in dataset.iterrows():
 if(2020 - row['Año'] != 0):
  dataset.loc[index, 'Km_media'] = row['Kilometraje'] / (2020 - row['Año'])
 else:
  dataset.loc[index, 'Km_media'] = 0
dataset
```

```python
#Itertuples() tambien es un iterador de lineas de un DataFrame retornando tuplas de valores
for row in dataset.itertuples():
 if(2020 - row.Año != 0):
  dataset.loc[row.Index, 'Km_media'] = row.Kilometraje / (2020 - row.Año)
 else:
  dataset.loc[row.Index, 'Km_media'] = 0
dataset
```
