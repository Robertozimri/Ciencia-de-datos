```python
#Fucniones que retornan valor  para esto utilizaremos return 
def media(lista):
  valor=sum(lista)/len(lista)
  return valor
resultado=media([1,2,3,4,5,6,7,8])
resultado
```

```python
#Funciona que retorna mas de un valor, estos retornan en general y vienen en forma de tuplas con dos o más valores
def media(lista):
  valor=sum(lista)/len(lista)
  return valor,len(lista)
media([1,2,3,4,5,6,7,8,9])
```

```python
resultado, n = media([1, 2, 3, 4, 5, 6, 7, 8, 9])
print(resultado)
print(n)
```
