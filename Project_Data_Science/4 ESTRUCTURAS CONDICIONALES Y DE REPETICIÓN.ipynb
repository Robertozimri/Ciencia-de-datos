```python
#instrucciones continue e break
#Continue es utilizada para salta el bloque de instrucciones actual y retorna la intruccion  y aqui busca todas las palabras con s
accesorios = ['Llantas de aleación', 'Cerraduras eléctricas', 'Piloto automático', 'Bancos de cuero',
'Aire acondicionado', 'Sensor de estacionamiento', 'Sensor crepuscular', 'Sensor de lluvia']
for item in accesorios:
 if(item[0] != 'S'):
   continue
 print(item)
```

```python
#Break es para salir totalmente de un bucle solo busca la primera palabra que empiece con S
for item in accesorios:
  if(item[0] !='S'):
    continue
  print(item)
  break
```

```python
#List Comprehensions [ <expresión> for <item> in <iterable> if <condición> ]
cuadrado = []
for i in range(10):
 cuadrado.append(i ** 2)
cuadrado
```

```python
#Lista comprehesions con clausulas if-elif-else 
datos = [
['Jetta Variant', 2003, False],
['Passat', 1991, False],
['Crossfox', 1990, False],
['DS5', 2019, True],
['Aston Martin DB4', 2006, False],
['Palio Weekend', 2017, False],
['A5', 2019, True],
['Série 3 Cabrio', 2009, False],
['Dodge Jorney', 2018, False],
['Carens', 2011, False]
]
```

```python
tipo=[]
for item in datos:
  if(item[1]>=2019):
    tipo.append('Vehiculo nuevo')
  else:
    tipo.append('Vehiculo usado')
tipo
```

```python
# Este es para utilizar IF-ELSE
['Vehiculo nuevo 'if (item[1]>=2019) else ('Vehiculo usado ') for item in datos]
```

```python
#Este para utilizar IF-ELSE-ELIF
tipo = []
for item in datos:
 if(item[1] >= 2019):
  tipo.append('Vehículo nuevo')
 elif(item[1] > 2016 and item[1] <= 2018):
  tipo.append('Vehículo seminuevo')
 else:
  tipo.append('Vehículo usado')
tipo
```

```python
['Vehiculo nuevo'  if(item[1] >= 2019) else 'Vehiculo seminuevo' if (item[1]>2016) and item[1]<=2018 else 'Vehiculo usado' for item in datos]
```
