```python
#6.11 Obteniendo estadisticas con arrays numpy 
#Estadisticas basicas para auxiliar en procesos de prueba y análisis
dataset = np.arange(1, 16).reshape((5, 3))
dataset
```

```python
#np.mean retorna la media de los elementos del array a lo largo del eje especificado
np.mean(dataset, axis=0)
```

```python
#Obtener la media por columna 
np.mean(dataset, axis=1)
```

```python
#np.std() retorna desviacion estándar de los elementos de array a lo largo del eje especificado
np.std(dataset, axis=0)
```

```python
np.std(dataset, axis=1)
```

```python
#np.sum() retorna la suma de los elementos de array a lo largo del eje especifico 
np.sum(dataset,axis=0)
```

```python
#Sumatoria por columna 
np.sum(dataset,axis=1)
```
