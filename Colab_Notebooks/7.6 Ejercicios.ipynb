```python
import pandas as pd
```

```python
proyecto=pd.read_csv('https://raw.githubusercontent.com/Robertozimri/Robertozimri/main/db_proyecto.csv', sep=';')
proyecto.head()
```

```python
proyecto.shape
```

```python
proyecto.drop_duplicates().shape
```

```python
proyecto.drop_duplicates(inplace=True)
proyecto.shape
```

```python
proyecto['ValorTotal'] = proyecto['Valor'] + proyecto['ValorMantenimiento']
proyecto.head()
```

```python
proyecto['ValorM2'] = proyecto['ValorTotal'] / proyecto['Area']
proyecto.head()
```

```python
proyecto['Caracteristicas'] = proyecto['Caracteristicas'].str[1:-1].str.replace("'", "").str.split(', ')
proyecto.head()
```

```python
for index, row in proyecto.iterrows():
    proyecto.loc[index,'Vigilancia']='Vigilancia 24 horas'in row['Caracteristicas']
proyecto.head()
```
